### Первая итерация градиентного спуска

```python
w_init = np.random.randint(-12, 10, 4) # начальные случайные веса, целые

# w_init = np.random.rand(4) - 4 случайных числа, нецелых
# w_init = np.random.randn(4) - 4 случайных числа из нормального распределения
```

```python
print(f'Начальные значения весов: {w_init}')
```

```python
print(f'Размерность вектора весов {w_init.shape} & Размерность матрицы объекты-признаки {X.shape}')
```

```python
predictions = X @ w_init # посчитали по предсказанию на каждый объект
print(f'Предскзания: {predictions}')
error = predictions - y_true # y_pred - y_true
print(f'Ошибка: {error}')
```

Предсказание для i-ого объекта:

$$
y_{\text{pred}_i} = model(x_i) = w_0 + \langle w, x_i\rangle = w_0 + w_1 * x_{i,1} + w_2 * x_{i,2} + ...
$$

Функция ошибки для i-ого объекта:

$$
\text{Loss}_i = (y_{\text{true}_i} - y_{\text{pred}_i}) ^ 2
$$

Функционал потерь - MSE:

$$
Q      = \text{mean}(\text{Loss}_i) = \text{mean}( (y_{\text{true}_i} - y_{\text{pred}_i}) ^ 2 ) = \text{mean}( (y_{\text{true}_i} - w_0 - w_1 * x_{i,1} - w_2 * x_{i,2} )^ 2)
$$

Частная производная функционала потерь по первому весу:

$$
\frac{\partial Q}{\partial w_1} = (\frac{1}{n} \cdot \sum (y_{\text{true}_i} - y_{\text{pred}_i}) ^ 2) )' = \frac{2}{n} \cdot \sum(y_{\text{true}_i} - y_{\text{pred}_i}) \cdot x_{i1}
$$

```python
# y_pred_i = model(x_i) = w_0 + <w, x_i> = w_0 + w_1 * x_i1 + w_2 * x_i2 + ...
# Loss_i = (y_true_i - y_pred_i) ^ 2
# Q      = mean(Loss_i) = mean( (y_true_i - y_pred_i) ^ 2 ) = mean( (y_true_i - (w_0 + w_1 * x_i1 + ...) )^ 2 )

# dQ / dw_1 = (1/n * sum((y_true_i - y_pred_i) ^ 2) )' = 2 / n * sum( (y_true_i - y_pred_i) * x_i1)
```

```python
MSE = np.mean(error ** 2) # (1/n) * sum( (y_true_i - y_pred_i) ^ 2 )
print(f'Начальное значение функционала потерь: {MSE}')
```

Первый вес: 
$$
2 *\cdot (y_{\text{pred}_0} - y_{\text{true}_0}) \cdot X_{0, 0}
$$

Второй вес: 
$$
2 *\cdot (y_{\text{pred}_0} - y_{\text{true}_0}) \cdot X_{0, 1}
$$

Третий вес: 
$$
2 *\cdot (y_{\text{pred}_0} - y_{\text{true}_0}) \cdot X_{0, 2}
$$

Нулевой вес: 
$$
2 *\cdot (y_{\text{pred}_0} - y_{\text{true}_0}) \cdot 1
$$


```python
'''
частная производная для функционала потерь по весам
np.mean(2 * (predictions - y_true) * X[:, 0]) - первый вес
np.mean(2 * (predictions - y_true) * X[:, 1]) - второй вес
np.mean(2 * (predictions - y_true) * X[:, 2]) - третий вес
np.mean(2 * (predictions - y_true) * X[:, 3]) - нулевой вес

== градиенту
'''
```

```python
gradient = (2 * X.T @ (predictions - y_true)) / len(y_true)
```

### 3. Gradient Descent

```python
import matplotlib.pyplot as plt

# X - матрица объектов-признаков
# y_true - правильные ответы

w = np.random.randint(-12, 10, 4)
learning_rate = 0.0001 

#fig, ax = plt.subplots(figsize=(18, 10))

final_loss = {}
for learning_rate in np.logspace(-5, -3, 15):
    for iteration in range(50):
        predictions = X @ w
        error       = predictions - y_true
        MSE         = np.mean(error ** 2)

        gradient    =  (2 * X.T @ error) / len(y_true)
        w           = w - gradient * learning_rate

        #print(f'Значение MSE после {iteration} итерации: {round(MSE)}')

    final_loss[learning_rate] = np.sqrt(MSE)
#plt.plot(np.arange(50), loss_history, label=f'lr: {learning_rate}')
#plt.legend()
#plt.show()
```
